```python

import random

name = input("Name : ")
print("Welcome " + name)

def get_word():
    words = ["apple", "alone", "america", "animal", "army", "art", "away", "baby", "bad", "bear", "beautiful", "black", "blue", "body", "book", "box", "brother", "brown", "captain", "cat", "century", "chart", "chief", "child", "church", "cloth", "cloud", "coast", "dark", "day", "develope", "different", "difficult", "doctor", "dog", "dress", "east", "earth", "egg", "energy", "England", "English", "equal", "europe", "face", "family", "fast", "fight", "food", "france", "game", "garden", "girl", "god", "gold", "glass", "green", "hair", "happy", "hard", "help", "home", "ice", "information", "interest", "island", "jÄ±b", "japanese", "key", "kind", "lady", "large", "language", "last", "laugh", "lead", "left", "level", "light", "line", "list", "loud", "machine", "main", "major", "maybe", "measure", "meat", "member", "minute", "molecules", "modern", "money", "month", "moon", "music", "never", "near", "nation", "name", "new", "next", "north", "note", "object", "ocean", "old", "on", "only", "open", "once", "opposite", "order", "oxygen", "page", "paint", "paper", "park", "party", "people", "period", "picture", "plant", "power", "question", "quiet", "race", "reach", "ready", "red", "rich", "safe", "scale", "sea", "score", "seat", "second", "similar", "small", "speed", "student", "tall"]
    word = random.choice(words)
    return word.upper()

def play(word):
    word_display = "_" * len(word)
    guesse = False
    predicted_letters = []
    predicted_words = []
    tries = 5
    
    print("Are you ready ?")
    print(word_display)
    print("\n")
    
    while not guesse and tries > 0:
        guess = input("Enter a letter : ").upper()  
        
        if len(guess) == 1 and guess.isalpha():
            if guess in predicted_letters:
                print("You tried the letter %s before!" %(guess))
            elif guess not in word:
                tries -= 1
                print("The letter {} is not in the word. You have {} rights left.".format(guess, tries))
                predicted_letters.append(guess)
            else:
                print("The letter %s is present in the word." %(guess))
                predicted_letters.append(guess)
                word_array = list(word_display)
                indices = [i for i, letter in enumerate(word) if  letter == guess]
                for index in indices: 
                    word_array[index] = guess
                word_display = "".join(word_array)
                if "_" not in word_display:
                    guesse = True
        elif len(guess) == len(word) and guess.isalpha():
            if guess == word:
                guesse = True
                word_display = word
            elif guess in predicted_words:
                print("%s you tried this word before" %(guess))
            else:
                print("Not the right answer!")
                predicted_words.append(guess)
                tries -= 1
                
        else:
            print("Invalid guess!")
        print(word_display)
        print("\n")
    
    
    if guesse:
        print("Congrats! You win.")
    else:
        print("Oh no! You didn't win this time.Correct answer : %s " %(word))
        
def main():
    word = get_word()
    play(word)
    while input("Do you want to try again? (Y/N)").upper() == "E":
        word = get_word()
        play(word)
main()
```
